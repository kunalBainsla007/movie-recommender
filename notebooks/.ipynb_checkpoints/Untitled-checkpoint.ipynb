```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/movies.csv')
df
```

```python
genre_counts = df[['action', 'scifi', 'drama', 'romance', 'thriller']].sum()
genre_counts.plot(kind='bar', color='steelblue')
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```
